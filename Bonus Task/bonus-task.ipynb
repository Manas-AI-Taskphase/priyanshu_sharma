### Importing Libraries

```python
import pandas as pd
import numpy as np
# import seaborn as sns
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
data = pd.read_csv("dinos.csv")
data.head()
```

Plotting the relationship of weight with respect to  different features

```python
data_num = data.drop(columns=["Species"])

fig, axes = plt.subplots(len(data_num.columns)//3, 3, figsize=(15, 6))
i = 0
for triaxis in axes:
    for axis in triaxis:
        data_num.hist(column = data_num.columns[i], ax=axis)
        i = i+1
```

### Mapping Values of Species and Gender to numerical features for Feature Scaling

```python
data = data.replace(to_replace= "M", value= 1).replace(to_replace= "F", value= 0)
data = data.replace(to_replace= "Iguanadon", value= 1).replace(to_replace= "Pterodactyl", value= 0).replace(to_replace= "Brachiosaurus", value= 0).replace(to_replace= "Raptor", value= 0).replace(to_replace= "Vulcanodon", value= 0).replace(to_replace= "Stegosaurus", value= 0)
mapping = {'Iguanadon': 0, 'Pterodactyl': 1, 'Brachiosaurus': 2, 'Raptor': 3, 'Vulcanodon': 4, 'Plateosaurus': 5, 'Stegosaurus': 6} 
data = data.replace({'Species': mapping})
```

### Feature Scaling

```python
mean = data.mean()
std = data.std()
data_set = (data - mean) / std
data_set
```
